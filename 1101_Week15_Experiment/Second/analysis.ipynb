```python
import os 
import plotly.graph_objects as go
import numpy as np
from scipy.signal import find_peaks
from scipy import stats
```

```python
lmbda= 650*10**(-9)
def smooth(data , mask):
    return [sum(data[i:i+mask])/mask for i in range(len(data)-mask-1)]
def inteference(x,I_0,d,a,distance):
    sin_theta = abs(indexToM(x)/distance)
    return I_0*(np.cos((np.pi*d*sin_theta)/lmbda)**2*(np.sin((np.pi*a*sin_theta)/lmbda)/((np.pi*a*sin_theta)/lmbda))**2)
def indexToM(index):
    return 8*10**(-3)*(index/6400)
def different (data):
    diff = [data[i+1]-data[i] for i in range(len(data)-1)]
    return (sum(diff)/len(diff),stats.sem(diff))
def dis (d):
    return lmbda*3.2/d 
def dis_1 (L):
    return lmbda*L/(0.25*10**(-3))
```
