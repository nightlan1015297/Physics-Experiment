```python
import random
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
possible_step = [-1,1]
class walker():
    def __init__(self):
        self.x = 0
        self.y = 0
        self.z = 0
        self.trajectory = {'x':[0],'y':[0],'z':[0]}
    
    def get_position(self):
        return [self.x,self.y,self.z]
    
    # Get the displacement from current point to the origin. 
    def get_displacement(self):
        return math.sqrt(self.x**2+self.y**2+self.z**2)

    def decision(self):
        if random.random()>0.6:
            return True
        else:
            return False
    def step_foward(self):
        self.x = self.x + possible_step[self.decision()]
        self.trajectory['x'].append(self.x)
        self.y = self.y + possible_step[self.decision()]
        self.trajectory['y'].append(self.y)
        self.z = self.z + possible_step[self.decision()]
        self.trajectory['z'].append(self.z)
    
    def get_x(self):
        return self.x
    
    def get_y(self):
        return self.y
        
    def get_z(self):
        return self.z

def get_mean_square(position_lis ,avg):
    sum = 0
    for i in position_lis:
        sum = sum + (i-avg)**2
    return sum/len(position_lis)
        
```

```python
step = [0]

avg_position = [0]
MS_position = [0]

walkers = [ walker() for i in range(10000)]

for s in range(1000):
    step.append( s + 1)
    position_sum = 0
    position_lis = []
    for element in walkers:
        element.step_foward()
        position_lis.append(element.get_x())
        position_sum += element.get_x()
    avg = position_sum/len(walkers)
    MS_position.append(get_mean_square(position_lis,avg))
    avg_position.append(avg)
```

```python
plt.plot(step,avg_position)
plt.show()
```

```python
plt.plot(step,MS_position)
plt.show()
```

```python
walkers = [ walker() for i in range(10000)]
step = [0]
avg_position = [0]
square_position = [0]
for _ in range(1000):
    step.append( _ + 1)
    position_x_sum = 0
    position_y_sum = 0
    for element in walkers:
        element.step_foward()
        position_x_sum += element.get_x()
        position_y_sum += element.get_y()
    square_position.append((position_x_sum**2+position_y_sum**2)/len(walkers))
    avg_position.append(math.sqrt(position_x_sum**2+position_y_sum**2)/len(walkers))
```
